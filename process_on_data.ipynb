# ERA 5 dataset

```python
import xarray as xr
import pandas as pd

# Define file path
file1_path = "/content/dataset/MyDrive/dataset/weatherpatterns/data_stream-oper_stepType-accum.nc"
file2_path = "/content/dataset/MyDrive/dataset/weatherpatterns/data_stream-oper_stepType-instant.nc"
file3_path = "/content/dataset/MyDrive/dataset/weatherpatterns/data_stream-oper_stepType-instant1.nc"
file4_path = "/content/dataset/MyDrive/dataset/weatherpatterns/data_stream-oper_stepType-instant2.nc"
file5_path = "/content/dataset/MyDrive/dataset/weatherpatterns/data_stream-oper_stepType-instant3.nc"
file6_path = "/content/dataset/MyDrive/dataset/weatherpatterns/data_stream-oper_stepType-accum1.nc"


# Load the NetCDF dataset
ds = xr.open_dataset(file1_path)
ds2 = xr.open_dataset(file2_path)
ds3 = xr.open_dataset(file3_path)
ds4 = xr.open_dataset(file4_path)
ds5 = xr.open_dataset(file5_path)
ds6 = xr.open_dataset(file6_path)



# Convert dataset to DataFrame
df = ds.to_dataframe().reset_index()
df2 = ds2.to_dataframe().reset_index()
df3 = ds3.to_dataframe().reset_index()
df4 = ds4.to_dataframe().reset_index()
df5 = ds5.to_dataframe().reset_index()
df6 = ds6.to_dataframe().reset_index()
```

# Merge DataFrames

```python
df6 = pd.concat([df6, df], ignore_index=True)

df6
```

```python
df2 = pd.concat([df2, df4], ignore_index=True)
df2
```

```python
# Merge based on common keys
df_wind = df3
df_geopotential = df5


# Merge based on common keys
df_surface = pd.merge(df6, df2, on=["valid_time", "latitude", "longitude", "number", "expver"], how="outer")


# Merge surface-level data with pressure-level data
df_final_wind = pd.merge(df_surface, df_wind, on=["valid_time", "latitude", "longitude", "number", "expver"], how="outer")
df_final_geopotential = pd.merge(df_surface, df_geopotential, on=["valid_time", "latitude", "longitude", "number", "expver"], how="outer")
```

```python
df_final_wind = df_final_wind.dropna(subset=['pressure_level'])
df_final_wind
```

```python
df_final_geopotential = df_final_geopotential.dropna(subset=['pressure_level'])
df_final_geopotential
```

```python
df_final_wind.to_csv("/content/dataset/MyDrive/dataset/wind_data_set.csv", index=False)
df_final_geopotential.to_csv("/content/dataset/MyDrive/dataset/geopotential_dataset.csv", index=False)
```
